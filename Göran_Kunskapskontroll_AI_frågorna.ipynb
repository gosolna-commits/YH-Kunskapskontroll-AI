1. AI, ML, DL
AI: artificiell intelligens är vetenskap om hur man kan efterlikna mänsklig intelligens.
ML: Maskin learning är en delmängd inom AI och syftar till att ta fram modeller för prediktion.
DL: Deep learning är en del av ML och innefattar neurala nätverk. Modeller för mönsterigenkänning. Påminner om hjärnans funktion med många samverkande beräkningsenheter.

2. Utvärderingsmått
Användning av K-delad korsvalidering används när man saknar valideringsdata. Bra vid små datavolymer. För varje körning t ex 5 st, beräknas RMSE och medelvärdet för dessa används som utvärderingsmått. 

3. Regressionsproblem innebär att utifrån givna x värden prediktera kontinuerliga y värden. Modeller som används för detta är t ex Linjär regression och beslutsträd. Tillämpningar kan t ex vara uppskattning av huspriser utifrån givna "features". Beräkning av bilpriser beroende på ett antal olika faktorer är ett annat exempel. Tillämpningsområdena är många. Inom medicin avseende blodtryck, dosering etc är sådana exempel.

5. RMSE tolkning
Ger ett värde som visar avvikelse mellan faktiskt värde och predikterat värde. Med differensen i kvardrat försvinner negativa värden. RMSE förstärker stora avvikelser. Om man bara önskar rangordna olika modeller kan MSE vara bättre. RMSE är mer resurskrävande. Lågt värde, små avvikelser, på RMSE är önskvärt

6. Klassificeringsproblem
Prediktion av kategorisk data där datan kan anta två eller flera klasser. Ej kontinuerliga värden.
Sjukdom och diagnos
Bildigenkänning.
Försäljning, kundkategorier.
Aktuella modeller
Logistisk regression,Support vector machines, Beslutsträd, Ensamble learning, Random forest

7. Confusion matrix.
En matris som visar hur bra en modell presterar. Jämför sanna och predikterade värden.
Exempel vid binär klassificering med Ja/nej. Antal gånger prdeikterat värde stämmer för varje Ja (True positiv) och Nej (True negative) samt motsvarande för felaktiga värden.

8. Precision
Precision beräknar andelen av positiva predikterade värden som är korrekt dvs hur ofta modellen har rätt svar.


9. Recall.
Mäter andelen av den positiva klassen som predikteras korrekt. Med andra ord hur många av alla sanna värden hittar modellen.


10. Ordinal, nominal.
Ordinal innebär att det ska finnas en naturlig ordning mellan kategorierna. Exempel betyg låg, medel och hög. Färger har ingen inbördes ordning. Sen är det en smaksak om "röd" är vackrast. Vid tilldelning av subjektiv rangordning kan färgerna bil ordinal. I så fall har Julia rätt.


11. Parametrar och hyperparametrar.
Hyperparametrar används för att styra hur vald modell ska tränas och fungera. Med Grid search kan man få hjälp att optimalt hitta lämpliga hyperparametrar.
Parametrar är värden som modellen lär sig under träning.
Hyperparamter t ex antal epochs, max depth, regularisering är värden som styr hur modellen ska lära sig.


12. Overfitting och underfitting
En överanpassad modell är mycket nära sanningen jämfört med träningsdata. Risken är då stor att den predikterar dåligt mot validerings och testdata.
En underanpassad modell predikterar dåligt i alla avseenden.

Overfitting kan åtgärdas med t ex
Mer träningsdata
Regularisering
Minska modellen komplexitet

Underfitting:
Välj bättre modell
Ändra optimerare
Öka antalet epochs
 


13. OvO och OvR algoritmerna
Två algoritmer för binär eller multiclass prediktion.
One vs Rest. Tillhör datapunkten en viss klass eller inte. Högst score sannolikt rätt klass.
One vs One är en algoritm där en datapunkt jämförs med var och en av de andra klasserna. Är datapunkten en 2:a jämfört med modellen 2->3, eller 2->4 osv. 

14. Regularisering av modell
Modellen begränsas för att motverka överanpassning / komlexitet. Exempel på regulariseringstekniker är l1,l2, drop out och early stopping. 

15. Bias Variance Trade-off
Bias underanpassad modell beroende på felaktigt antagande om linjär/polynomregression.
Variance, Överkänslig modell för fluktueringar i datan. Överanpassad modell.
Irreducerbart fel beror på att datan är slumpmässig och inte passar till modellen.

16. Komplexa modeller, fördelar och nackdelar
Komplexa modeller är ofta mycket anpassad till träningsdata. Risken är då stor att modellen inte presterar bra i andra sammanhang. 
Komplexa modeller tenderar också att kräva mycket mycket mer tid och maskinresurser vid träningen. 

17. Skattning av okända parametrar. 
Söker efter variabelvärden som minimerar MSE. Gradient descent är en optimeringsalgoritm som letar linjens minsta MSE/ loss. I minimipunktern t ex ett globalt minimum" är linjens lutning 0. Där är MSE värdet lägst.

18. MSE eller RMSE
Man söker bara minsta väre för avvikelse. MSE kan gärna användas för jämförelser och utvärdering av modeller. Nackdelen är att felet presenteras i "kvadrat"
Om värdet är relevant att se på så fungerar inte MSE med t ex lön. Konstiga siffror. Med RMSE tar man kvadratroten ur MSE och siffrorna blir med relevant att se på.
Om man använder MSE kan man spara på maskinresurser eftersom man undviker beräkning av kvadratrot.

19. Övervakad och oövervakad maskininlärning
Vid linjär regression kan modellen lära sig av y-parametrarna som är kontinuerliga värden. Svaret är känt från början. Regression/Klassifisering.
Vid oövervakad inlärning används bara x-värden för träning av modellen. Klustring och dimensionsreducering.

20. Prediktions / konfidensintervall.
Mätning av osäkerhet. 
Konfidensintervallet är ett intervall med medelvärden. Ett givet värde ligger därmed med största säkerhet inom detta intervall. Gäller om medelvärdet är korrekt.
Prediktionsintervallet är mer osäkert beroende på flera faktorer som korrekt medelvärde samt variation mellan olika observationer. Enskilda värden varierar mycket mer än medelvärden. Prediktionsintervallet blir i och med det bredare.
